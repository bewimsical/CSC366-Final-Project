```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlencode
from google.colab import userdata

max_pages_to_search = 10

list_of_urls = [
    ['desktops','https://www.newegg.com/p/pl?N=100019095&d=desktops&PageSize=96&Order=3&page='],
    ['laptops' ,'https://www.newegg.com/p/pl?N=100006740&d=laptops&PageSize=96&Order=3&isdeptsrh=1&page='],
    ['phones'  ,'https://www.newegg.com/p/pl?d=phone&PageSize=96&Order=3&page=']
]

new_egg_soup = {}

# Should end up like this
# new_egg_soup = {
#     desktops: {
#         1: "",
#         2: "",
#         etc...
#     },
#     laptops: {
#         1: "",
#         2: "",
#         etc...
#     },
#     phones:{
#         1: "",
#         2: "",
#         etc...
#     }
# }


# for each url
# go through each page e.i. page=1, page=2, etc
for (product_type, url) in list_of_urls:
  new_egg_soup[product_type] = {}
  for page in range(max_pages_to_search):
    print(f"Scanning Page {page+1} of '{product_type}'...\n{url}{page+1}")
    params = {'api_key': userdata.get('scraperAPI'), 'url': f"{url}{page+1}"}
    response = requests.get('http://api.scraperapi.com/', params=urlencode(params))

    soup = BeautifulSoup(response.text, 'html.parser')

    product_list = soup.find('div',class_='item-cells-wrap') # get the product list

    new_egg_soup[product_type][page+1] = product_list.contents # save all childern
```

```python
# get the first desktop on page 1 of 'desktops'
new_egg_soup['desktops'][1][0]
```

```python
# Parse all relevant information
for product_type in new_egg_soup:
  for page in new_egg_soup[product_type]:
    for product in new_egg_soup[product_type][page]:
      # print(f"{product}\n\n\n")

      name = ""
      rating = ""
      review_count = ""
      model_number = ""
      price = ""
      brand = ""

      # parse name/title
      name_el = product.find("a", class_="item-title")
      if (name_el): name = name_el.text

      # parse the rating
      review_el = product.find('a',class_='item-rating')
      if (review_el):
        rating = review_el.attrs['title'].replace("Rating + ", "")
        # parse the rating count
        review_count = review_el.find('span').text.strip('()')

      # parse the model number
      model_tag = product.find("strong", string="Model #: ")
      if model_tag and model_tag.parent:
          model_number = model_tag.parent.text.replace("Model #: ", "")

      # parse price
      price_el = product.find("li", class_="price-current")
      if (price_el): price = price_el.text.replace("–", "")

      # parse brand
      brand_el = product.find("a", class_="item-brand")
      if (brand_el):
        brand_img_el = brand_el.find("img")
        if (brand_img_el): brand = brand_img_el.attrs['title']

      print(f"\
        NAME: {name}\n\
        RATING: {rating or None}\
        REVIEW COUNT: {review_count or None}\
        MODEL #: {model_number or None}\
        PRICE: {price or None}\
        BRAND: {brand or None}\
        \n\
      ")
```

```python
# Need to add secret key into Colab.
# This takes a couple minutes to run.

import re
import pyjson5
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlencode
from google.colab import userdata

max_pages_to_search = 10

list_of_urls = [
    ['desktops','https://www.newegg.com/p/pl?N=100019095&d=desktops&PageSize=96&Order=3&page='],
    ['laptops' ,'https://www.newegg.com/p/pl?N=100006740&d=laptops&PageSize=96&Order=3&isdeptsrh=1&page='],
    ['phones'  ,'https://www.newegg.com/p/pl?d=phone&PageSize=96&Order=3&page=']
]

egg_soup = {}

for (product_type, base_url) in list_of_urls:
    egg_soup[product_type] = {}

    for page in range(max_pages_to_search):
        full_url = f"{base_url}{page + 1}"
        print(f"Scanning Page {page + 1} of '{product_type}'...\n{full_url}")

        response = requests.get('http://api.scraperapi.com/', params = {'api_key': userdata.get('scraperAPI'), 'url': full_url})
        html = response.text

        # Parse HTML with BeautifulSoup
        soup = BeautifulSoup(html, 'html.parser')

        # Find the <script> tag containing window.__initialState__
        script_tag = soup.find("script", string=re.compile("window.__initialState__"))
        if not script_tag:
            print(f"No script tag found on page {page + 1}")
            continue

        script_text = script_tag.string or script_tag.text
        if not script_text:
            print(f"Script tag empty on page {page + 1}")
            continue

        # Extract the JSON substring
        start = script_text.find("window.__initialState__ = ") + len("window.__initialState__ = ")
        end = script_text.find(";</script>", start)
        if end == -1:
            end = len(script_text)

        json_text = script_text[start:end].strip().rstrip(";")

        try:
            parsed_json = pyjson5.loads(json_text)
            egg_soup[product_type][page + 1] = parsed_json
            print(f"Parsed JSON successfully for {product_type} page {page + 1}")
        except Exception as e:
            print(f"Error parsing JSON for {product_type} page {page + 1}: {e}")

```

```python

for product_type in egg_soup:
  for page in egg_soup[product_type]:
    for product in egg_soup[product_type][page]['Products']:

      name = product['ItemCell']['Description']['Title']
      rating = product['ItemCell']['Review']['RatingOneDecimal']
      review_count = product['ItemCell']['Review']['HumanRating']
      model_number = product['ItemCell']['Model']
      price = product['ItemCell']['FinalPrice']
      brand = product['ItemCell']['ItemManufactory']['Manufactory']

      print(f"\
        NAME: {name}\n\
        RATING: {rating or None}\
        REVIEW COUNT: {review_count or None}\
        MODEL #: {model_number or None}\
        PRICE: {price or None}\
        BRAND: {brand or None}\
        \n\
      ")

```
