```python


# Get all JSON files in the folder-change the folder to merge different groups of files
file_paths = glob.glob("laptop/*.json")

# Load all product lists
all_products = []

for path in file_paths:
    with open(path, "r") as file:
        data = json.load(file)
        # assuming each file contains a list of products
        all_products.extend(data["products"])

# Save the combined list into one JSON file
with open("all-laptop-data.json", "w") as file:
    json.dump(all_products, file, indent=4)

print(f"Merged {len(file_paths)} files, total {len(all_products)} products.")
```
